```python
#Let us create a function for the execution of these codes
def find_jobs():
    jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')
    i = 1 #(Using this for count purpose in the result)
    # using for loop for checking all the result
    for job in jobs:
        published_date = job.find('span', class_ = 'sim-posted').text

    # we are checking if the word 'few' is inside the published_date
        if 'few' in published_date:
            company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(" ","")
            skills = job.find('span' , class_ = 'srp-skills').text.replace(" ","")
            if familiar_skill in skills:
                #header is used in more info as the first tag because the link lies in the heading h2
                # a is the tag for the link and index we have is 'href'
                more_info = job.header.h2.a['href']
                print(i)
                print(" ")
                print(f'Company Name: {company_name}')
                print(f'Required Skills: {skills}')
                print(f'Published Date: {published_date}')
                print(f'More info: {more_info}')
                # printing some space at the end
                print(" ")
                i = i + 1
```

```python
# to automate the execution of code as per need

import time
if __name__ == '__main__':
    while True:
        find_jobs()
        time_wait = 10
        print(f'Waiting time {time_wait} minutes...')
        time.sleep(time_wait*60)
```

```python
# getting the results into a text file

def find_jobs():
    html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Analyst&txtLocation=India&cboWorkExp1=0').text
    soup = BeautifulSoup(html_text, 'lxml')
    jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')
    i = 1 #(Using this for count purpose in the result)
    # using for loop for checking all the result
    for index, job in enumerate(jobs):
        published_date = job.find('span', class_ = 'sim-posted').text

    # we are checking if the word 'few' is inside the published_date
        if 'few' in published_date:
            company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(" ","")
            skills = job.find('span' , class_ = 'srp-skills').text.replace(" ","")
            if familiar_skill in skills:
                with open(f'posts/{index}.txt', 'w') as f:
                    #header is used in more info as the first tag because the link lies in the heading h2
                    # a is the tag for the link and index we have is 'href'
                    more_info = job.header.h2.a['href']
                    f.write(i)
                    f.write(" ")
                    f.write(f'Company Name: {company_name} \n')
                    f.write(f'Required Skills: {skills} \n')
                    f.write(f'Published Date: {published_date} \n')
                    f.write(f'More info: {more_info} \n')
                    # printing some space at the end
                    f.write(" ")
                    i = i + 1
                 print(f'file saved: {index}')   

                
# to automate the execution of code as per need

import time
if __name__ == '__main__':
    while True:
        find_jobs()
        time_wait = 10
        print(f'Waiting time {time_wait} minutes...')
        time.sleep(time_wait*60)                
```
